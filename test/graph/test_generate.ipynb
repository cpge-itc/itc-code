## Unweighted

### Undirected

### Directed

```python
G = cpge.graph.random_matrix(directed=True)
cpge.graph.draw(G, directed=True)
```

```python
G = cpge.graph.random_list(directed=True)
cpge.graph.draw(G, directed=True)
G
```

## Weighted

### Undirected

```python
G = cpge.graph.random_matrix(weighted=True, directed=False)
cpge.graph.draw(G, directed=False, weighted=True)
```

```python
G = cpge.graph.random_list(weighted=True, directed=False)
cpge.graph.draw(G, directed=False, weighted=True)
```

### Directed


```python
G = cpge.graph.random_matrix(weighted=True, directed=True, p=0.2)
cpge.graph.draw(G, directed=True, weighted=True)
```

```python
G = cpge.graph.random_list(weighted=True, directed=True, p=0.2)
cpge.graph.draw(G, directed=True, weighted=True)
```
