## Unweighted

### Undirected

```python
G = itc.graph.random_matrix()
itc.graph.draw(G)
```

```python
G = itc.graph.random_list()
itc.graph.draw(G)
G
```

### Directed

```python
G = itc.graph.random_matrix(directed=True)
itc.graph.draw(G, directed=True)
```

```python
G = itc.graph.random_list(directed=True)
itc.graph.draw(G, directed=True)
G
```

## Weighted

### Undirected

```python
G = itc.graph.random_matrix(weighted=True, directed=False)
itc.graph.draw(G, directed=False, weighted=True)
```

```python
G = itc.graph.random_list(weighted=True, directed=False)
itc.graph.draw(G, directed=False, weighted=True)
```

### Directed


```python
G = itc.graph.random_matrix(weighted=True, directed=True, p=0.2)
itc.graph.draw(G, directed=True, weighted=True)
```

```python
G = itc.graph.random_list(weighted=True, directed=True, p=0.2)
itc.graph.draw(G, directed=True, weighted=True)
```
